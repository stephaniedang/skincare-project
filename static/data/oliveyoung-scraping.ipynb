```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# URL = "productList.html"
# page = requests.get(URL)

soup = BeautifulSoup(open("productList.html"), "html.parser")
print(soup.prettify())
```

```python
results = soup.find_all(class_="unit-desc")
print(results)

products = soup.findAll("dd")
print(products)

prices = soup.findAll(class_="price-info")
print(prices)

ratings = soup.findAll(class_="rating-info")
print(ratings)
```

```python
df = pd.DataFrame(columns=["productName", "price", "rating"])
rows = []

for result in results:
  product = result.find('dd').text
  # print(product)

  price_info = list(filter(None, result.find('div', attrs={'class':'price-info'}).text.split('\n')))
  if len(price_info) == 1:
    price = price_info[0]
  elif len(price_info) == 2:
    price = price_info[1]
  else:
    price = "Price information not available"
  # print(price)

  rating = result.find('div', attrs={'class':'rating-info'}).text.split('\n')[1]
  # print(rating)

  rows.append({"productName": product, "price": price, "rating": rating})

df = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)
```
